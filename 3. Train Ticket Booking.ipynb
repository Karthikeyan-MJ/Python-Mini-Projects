```python
import pandas as pd
```

```python
ALL_BOOKINGS = []
Stations = ['chennai', 'mumbai', 'banglore', 'hyderabad', 'kerala', 'pune', 'gujarat', 'mysore', 'renigunta']

def Bookings():

    print('Total no. of tickets booked are')
    print('\n The passengers details are :-')
    booked = []
    booked.append(ALL_BOOKINGS)
    return print(booked)


def stations():

    station_names = {'Chennai': {'mumbai': '5.00am', 'banglore': '5.30am', 'kerala': '7.20am'}, 'mumbai': {'Chennai': '6.00am', 'Pune': '6.00am', 'Gujarat': '6.45am'}, 'banglore':
                     {'Hyderabad': '7.00am', 'Mysore': '5.45am', 'Chennai': '7.00am'}, 'hyderabad': {'Banglore': '8.00am', 'Renigunta': '7.45am', 'Mysore': '8.45am'}}

    print("\nThe The trains available in each stations are displayed above here :-")
    for key, values in station_names.items():
        print('\nIn {} Station:- available trains are {} '.format(key, values))
          
    station = pd.DataFrame({'Stations':['Chennai','Chennai','Chennai', 'mumbai', 'mumbai', 'mumbai', 'banglore', 'banglore', 'banglore', 'hyderabad', 'hyderabad', 'hyderabad'], 'Trains_to': ['mumbai', 'Banglore', 'kerala', 'chennai', 'pune', 'gujarat', 'hyderabad', 'mysore', 'chennai', 'banglore', 'renigunta', 'mysore'], 'Timing': ['5.00am', '5.30am', '7.20am', '6.00am', '6.00am', '6.45am', '7.00am', '5.45am', '7.00am', '8.00am', '7.45am', '8.45am']})
    
    return print(station)


def Schedules():
    print('\nThe Shedules for Availability of trains')
    print('\nThe schedules in Chennai Stations are :-\n')
    Chennai_schedules = {'mumbai': '5.00am',
        'banglore': '5.30am', 'kerala': '7.20am'}
    for key, values in Chennai_schedules.items():
        print('Train to {} is starting at {}'.format(key, values))

    print('\nThe schedules in Mumbai Stations are :-\n')
    Mumbai_schedules = {'Chennai': '6.00am',
        'Pune': '6.00am', 'Gujarat': '6.45am'}
    for key, values in Mumbai_schedules.items():
        print('Train to {} is starting at {}'.format(key, values))

    print('\nThe schedules in Banglore Stations are :-\n')
    banglore_schedules = {'Hyderabad': '7.00am',
        'Mysore': '5.45am', 'Chennai': '7.00am'}
    for key, values in banglore_schedules.items():
        print('Train to {} is starting at {}'.format(key, values))

    print('\nThe schedules in Hyderabad Stations are :-\n')
    hyderabad_schedules = {'Banglore': '8.00am',
        'Renigunta': '7.45am', 'Mysore': '8.45am'}
    for key, values in hyderabad_schedules.items():
        print('Train to {} is starting at {}'.format(key, values))

    return


def Book_a_Ticket():

    Name = input('Enter your name :-')
    while not Name.isalpha():
        Name = input('Invalid input, Enter your name :-')

    Tickets = input('Enter no. of tickets :- ')
    while not Tickets.isnumeric():
        Tickets = input('Invalid Input, Enter no. of tickets :- ')

    From = input('Enter Travelling from :- ')
    while not From.isalpha():
        From = input('Invalid station, Enter Travelling from :- ')
    if From in Stations:
        pass
    else:
        From = input('Invalid station, Enter Travelling from :- ')

    To = input('Enter Travelling to :- ')
    while not To.isalpha():
        To = input('Invalid station, Enter Travelling to :- ')
    if To in Stations:
        pass
    else:
        To = input('Invalid station, Enter Travelling to :- ')

    booking = {'Name': Name,'No of Tickets': Tickets,'From': From,'To': To}

    print('\nThe passengers details :- ')
    for key, values in booking.items():
        print('\n{} :- {}'.format(key, values))

    ALL_BOOKINGS.append(booking)

    print('\nThe traveller name is {}, no. of tickets booked is {} are travelling from {} to {}.'.format(
        Name, Tickets, From, To))
    return


print('\nWelcome to Railway Booking System')
print('\nSelect an option below')

while(True):
    Menu = input('''\nTrain Ticket Booking System
    1. Bookings
    2. Stations
    3. Schedules
    4. Book a Ticket
    5.Exit

    Enter Option :- ''')
    while not Menu.isnumeric():
        Menu = input('Invalid input, Enter a Number :- ')

    Menu = int(Menu)
    if Menu == 1:
        Bookings()
    elif Menu == 2:
        stations()
    elif Menu == 3:
        Schedules()
    elif Menu == 4:
        Book_a_Ticket()
    elif Menu == 5:
        print('Closing the system')
        break
```
